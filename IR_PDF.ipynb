```python
import PyPDF2
from PyPDF2 import PdfReader
def extract_text_from_pdf(pdf_path):
    pdf = PdfReader(pdf_path)
    text = ''
    for page in pdf.pages:
        text += page.extract_text()
    return text

```

```python
def parse_faqs(text):
    faqs = {}
    lines = text.split('\n')
    question = None
    answer = ''
    for line in lines:
        if line.startswith('Q:'):
            if question and answer:  # save the previous question-answer pair
                faqs[question] = answer.strip()
            question = line[3:].strip()  # remove 'Q:' from the start of the question
            answer = ''  # start a new answer
        else:
            answer += line + '\n'  # add this line to the current answer
    if question and answer:  # save the last question-answer pair
        faqs[question] = answer.strip()
    return faqs
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

def find_most_similar_question(new_question, faq_questions):
    if not new_question or not faq_questions:
        return None
    vectorizer = TfidfVectorizer().fit_transform([new_question] + list(faq_questions))
    similarity_matrix = cosine_similarity(vectorizer[0:1], vectorizer[1:])
    most_similar_question_index = similarity_matrix.argmax()
    return list(faq_questions)[most_similar_question_index]
```

```python
def get_answer(question, faq_dict):
    return faq_dict[question]

```

```python
# usage
text = extract_text_from_pdf('/content/faq_G20_merged.pdf')
faq_dict = parse_faqs(text)
new_question = 'How to delete my EG2 account?'
similar_question = find_most_similar_question(new_question, faq_dict.keys())
answer = get_answer(similar_question, faq_dict)
print(answer)
```
